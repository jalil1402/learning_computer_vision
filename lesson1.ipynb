```python
import cv2 as cv
from google.colab.patches import cv2_imshow
image=cv.imread("/content/WIN_20240327_17_08_20_Pro.jpg")
cv2_imshow(image)
text = "Hello, World!"
font = cv.FONT_HERSHEY_SIMPLEX
font_scale = 1
font_color = (0, 255, 0)  # Green color in BGR
thickness = 2

# Get text size
text_size = cv.getTextSize(text, font, font_scale, thickness)[0]

# Calculate text position to center it
text_x = (image.shape[1] - text_size[0]) // 2
text_y = (image.shape[0] + text_size[1]) // 2

# Put text on the image
cv.putText(image, "jalil", (text_x, text_y), cv.FONT_HERSHEY_COMPLEX, 1, (0,255,0),2)

# Display the image with text
cv2_imshow(image)
cv.waitKey(0)
```

```python
# colerless
gray=cv.cvtColor(image,cv.COLOR_RGB2GRAY)
cv2_imshow(gray)
```

```python
#blur
blur=cv.GaussianBlur(image,(7,7),cv.BORDER_ISOLATED)
cv2_imshow(blur)
# edge
edge=cv.Canny(image,125,175)
cv2_imshow(edge)
```

```python
CONTOURS,HEIARCHIES=cv.findContours(edge,cv.RETR_EXTERNAL,cv.CHAIN_APPROX_SIMPLE)
import numpy as np

print(len(CONTOURS))
ret,thresh=cv.threshold(image,175,255,cv.THRESH_BINARY)
cv2_imshow(thresh)
# drawing the counters
blank=np.zeros(image.shape,"")
counters=cv.drawContours(blank,CONTOURS,-1,(255,0,0),2)
cv2_imshow(conters)
```
